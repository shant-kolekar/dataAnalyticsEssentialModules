## Crosstab | Contingency table

`pd.crosstab()`

The `pd.crosstab()` function is used to compute a cross-tabulation of two or more factors, essentially creating a contingency table.

**Syntax:**

```python
pd.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, margins_name='All', normalize=False)


**Common Parameters:**

- **`index`**: Series, array-like, or list of Series/arrays. The rows of the crosstab table.
- **`columns`**: Series, array-like, or list of Series/arrays. The columns of the crosstab table.
- **`values`**: Array of values to aggregate according to the factors (optional).
- **`aggfunc`**: Function to aggregate if values are provided, e.g., `sum`, `mean`, `min`, `max`, `count`, `std`, `var`.
- **`margins`**: If True, it will add margins (subtotals) to the rows and columns.
- **`margins_name`**: The name of the row/column that will contain the totals when `margins=True`. Default is 'All'.
- **`normalize`**: If True, normalizes by dividing all values by the sum of total values (can also be used for row/column normalization).


````python

import pandas as pd

# Sample data
data = {
    'Job Role': ['Engineer', 'Manager', 'Analyst', 'Engineer', 'Manager', 'Analyst', 'Engineer', 'Analyst'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'Finance', 'HR', 'Finance', 'IT'],
    'Salary': [70000, 90000, 50000, 75000, 95000, 52000, 72000, 55000]
}

df = pd.DataFrame(data)

# Crosstab with all common parameters
result = pd.crosstab(
    index=df['Job Role'],
    columns=df['Department'],
    values=df['Salary'],
    aggfunc='sum',
    margins=True,
    margins_name='Overall Sum',
    normalize=True
)

result
````

`df.groupby(['country']).size().reset_index(name='total_users')`

1. **`df.groupby(['country'])`:**
   - This groups the DataFrame by the `country` column.
   - The result is a special grouping object, where each unique `country` value represents a group.

2. **`.size()`:**
   - This counts how many rows exist in each group (i.e., for each country).
   - The result is a **Series** where:
     - The **index** is the unique values in the `country` column.
     - The **values** are the counts of rows in each group (how many users per country).

3. **Resulting Structure (Before `.reset_index()`):**
   - After `.size()`, you get a **Series** where:
     - **Index:** The country names.
     - **Values:** The count of users in each country.
   


  Example output:

   | Index   | 0    |
   |---------|------|
   | USA     | 120  |
   | Canada  | 80   |
   | Germany | 50   |

4. **`.reset_index(name='total_users')`:**
   - `reset_index()` converts the index (the `country` names) into a regular column in a new DataFrame.
   - The values (the counts) are placed in a new column.
   - By specifying `name='total_users'`, you are giving a name to this new column, which holds the counts.

5. **Final Structure (After `.reset_index()`):**
   - The result is a **DataFrame** with two columns:
     - One column is the original index (the country names).
     - Another column holds the counts, which is named `total_users`.

`df.merge()`

The `df.merge()` function in pandas is used to merge (or join) two DataFrames based on one or more keys. It is similar to SQL joins and allows you to combine datasets on common columns or indices.

 **Syntax:**

```python
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), indicator=False, validate=None)



**Common Parameters:**

- **`left`**: The left DataFrame.
- **`right`**: The right DataFrame.
- **`how`**: The type of merge to be performed. Options are:
  - `'left'`: Keep all keys from the left DataFrame.
  - `'right'`: Keep all keys from the right DataFrame.
  - `'outer'`: Keep all keys from both DataFrames.
  - `'inner'`: Keep only the keys that are common to both DataFrames (default).
- **`on`**: Column(s) to join on. Must be found in both DataFrames. If not specified, and no `left_on` or `right_on` are specified, it uses the intersection of columns in both DataFrames.
- **`left_on`**: Column(s) from the left DataFrame to join on.
- **`right_on`**: Column(s) from the right DataFrame to join on.
- **`left_index`**: If `True`, use the index from the left DataFrame as the join key.
- **`right_index`**: If `True`, use the index from the right DataFrame as the join key.
- **`suffixes`**: Suffix to apply to overlapping column names in the result. Default is `('_x', '_y')`.
- **`indicator`**: If `True`, adds a column to the output DataFrame called `_merge`, which indicates the source of each row. Values are:
  - `'left_only'`: Row only appears in the left DataFrame.
  - `'right_only'`: Row only appears in the right DataFrame.
  - `'both'`: Row appears in both DataFrames.
- **`sort`**: If `True`, sorts the resulting DataFrame by the join keys.
- **`validate`**: String indicating whether to check if the merge is of a specific type. Options are:
  - `'one_to_one'`: Validate that each row in the left DataFrame has only one corresponding row in the right DataFrame and vice versa.
  - `'one_to_many'`: Validate that one row in the left DataFrame can be matched to multiple rows in the right DataFrame.
  - `'many_to_one'`: Validate that many rows in the left DataFrame match one row in the right DataFrame.
  - `'many_to_many'`: Allows multiple matches between both DataFrames.



**Example**

pd.merge_ordered(left, right, on=None, left_on=None, right_on=None,

`pd.merge_ordered()`

The `pd.merge_ordered()` function is used to merge two DataFrames in a way that preserves the order of the keys.

* Common Parameters

  - **`left`**: The left DataFrame to merge.
  - **`right`**: The right DataFrame to merge.
  - **`how`**: The type of merge to be performed. Options include:
    - **`'left'`**: Keep all keys from the left DataFrame.
    - **`'right'`**: Keep all keys from the right DataFrame.
    - **`'outer'`**: Keep all keys from both DataFrames (default).
    - **`'inner'`**: Keep only the keys that are common to both DataFrames.
  - **`on`**: Column(s) to join on. Must be found in both DataFrames. If not specified and no `left_on` or `right_on` are specified, it uses the intersection of columns in both DataFrames.
  - **`left_on`**: Column(s) from the left DataFrame to join on.
  - **`right_on`**: Column(s) from the right DataFrame to join on.
  - **`left_index`**: If `True`, use the index from the left DataFrame as the join key.
  - **`right_index`**: If `True`, use the index from the right DataFrame as the join key.
  - **`suffixes`**: Suffixes to apply to overlapping column names in the result. Default is `('_x', '_y')`.
  - **`indicator`**: If `True`, adds a column called `_merge` to the output DataFrame, indicating the source of each row. Values can be:
    - **`'left_only'`**: Row only appears in the left DataFrame.
    - **`'right_only'`**: Row only appears in the right DataFrame.
    - **`'both'`**: Row appears in both DataFrames.
  - **`sort`**: If `True`, sorts the resulting DataFrame by the join keys.
  - **`validate`**: A string indicating whether to check if the merge is of a specific type. Options include:
    - **`'one_to_one'`**: Validate that each row in the left DataFrame has only one corresponding row in the right DataFrame and vice versa.
    - **`'one_to_many'`**: Validate that one row in the left DataFrame can be matched to multiple rows in the right DataFrame.
    - **`'many_to_one'`**: Validate that many rows in the left DataFrame match one row in the right DataFrame.
    - **`'many_to_many'`**: Allows multiple matches between both DataFrames.

    - **`fill_method`**: `ffill` or `bfill`

        * Forward Fill (`ffill`)
          - **Description**: Fills missing values by propagating the last valid observation forward.
        * Backward Fill (`bfill`)
          - Forward Fill (`ffill`)
          - **Description**: Fills missing values by propagating the last valid observation backward.

```python
import pandas as pd

# DataFrame 'a' with index as the time key
a = pd.DataFrame({'Value_A': [100, 200, 300]},
                 index=pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-04']))

# DataFrame 'b' with a 'Date' column
b = pd.DataFrame({'Date': pd.to_datetime(['2023-01-01', '2023-01-03', '2023-01-04']),
                  'Value_B': [10, 30, 40]})

# Merge ordered with no fill
result = pd.merge_ordered(a, b, left_index=True, right_on='Date')
print(result)

- #Merge ordered with no fill

    Value_A       Date  Value_B
0    100.0   2023-01-01     10
1    200.0   2023-01-02    NaN
2    NaN     2023-01-03     30
3    300.0   2023-01-04     40

# Merge ordered with forward fill
result = pd.merge_ordered(a, b, left_index=True, right_on='Date', fill_method='ffill')
print(result)

Forward Fill (ffill):
   Value_A       Date  Value_B
0    100.0 2023-01-01     10.0
1    200.0 2023-01-02      NaN
2    200.0 2023-01-03     30.0
3    300.0 2023-01-04     40.0


Backward Fill (bfill):
   Value_A       Date  Value_B
0    100.0 2023-01-01     10.0
1    300.0 2023-01-02      NaN
2    300.0 2023-01-03     30.0
3    300.0 2023-01-04     40.0


```

* In this case, the `NaN` in `Value_B` occurs because the Date `2023-01-02` is in DataFrame `a` but does not exist in DataFrame `b`. Since `Value_B` doesn't have any data for that date, the merge creates a `NaN` in that row. The `ffill` method only applies to gaps within DataFrame `b`, but there's no missing data in consecutive rows in `b` for ffill to act upon.

* How ffill Works in This Example:
DataFrame a (`Value_A` column): The `200` value from `2023-01-02` is forward-filled to `2023-01-03` because they are consecutive rows in `a`.
DataFrame `b` (`Value_B` column): There is no consecutive missing data, so nothing gets forward-filled in `Value_B`.






## Joining to nearest date-times

* Why Go for merge_asof() if we have pd.merg(). Well, If we wanted some data instead of NaN for non matching reocrds then this is the method to opt for.


`pd.merge_asof()`

The `pd.merge_asof()` function is used to perform a merge by searching for the nearest key rather than requiring exact matches. This is particularly useful for time series data.

> Common Parameters

- **`left`**: The left DataFrame to merge.
- **`right`**: The right DataFrame to merge.
- **`left_on`**: Column name(s) from the left DataFrame to merge on. Must be a sorted column.
- **`right_on`**: Column name(s) from the right DataFrame to merge on. Must also be sorted.
- **`by`**: Column name(s) to group by before merging. Useful for handling multiple categories.
- **`tolerance`**: Maximum allowed distance between the merge keys. If the distance exceeds this value, the merge will not be made.
- **`direction`**: Specifies which direction to search for the nearest match. Options are:
  - **`'backward'`**: Use the last value before the key.
  - **`'forward'`**: Use the first value after the key.
  - **`'nearest'`**: Use the nearest value, regardless of direction. Default is `'backward'`.
- **`suffixes`**: A tuple of strings to apply as suffixes to overlapping column names. Default is `('_x', '_y')`.

> Example

Heres an example demonstrating the usage of `pd.merge_asof()`:

```python
import pandas as pd

# Sample DataFrames
prices = pd.DataFrame({
    'date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-04']),
    'price': [100, 200, 300]
})

trades = pd.DataFrame({
    'trade_date': pd.to_datetime(['2023-01-01', '2023-01-03']),
    'trade_volume': [10, 15]
})

# Merging using merge_asof
merged = pd.merge_asof(trades, prices, left_on='trade_date', right_on='date', direction='backward')

print(merged)

   trade_date  trade_volume  price
0  2023-01-01            10    100
1  2023-01-03            15    200

```

> Explanation of the Example

* `DataFrames`: Two DataFrames are created: `prices`, which contains stock prices on specific dates, and `trades`, which contains trades with their corresponding dates.

* `Merge Operation`: The pd.merge_asof() function merges the two DataFrames on the trade dates, finding the nearest price for each trade date.

* `Direction`: The direction='backward' parameter ensures that the nearest price is taken from the last available price before the trade date.

* method: Defines how to handle same values (**Default**--->'average', 'min', 'max', 'first').
* axis: Specifies the axis for ranking (**Default**-->0 or 1).
* ascending: Defines the order (**Default**-->True for ascending, False for descending).
* na_position: Determines the rank position for NaN values ('first' or **Default**-->'last').



> Arg set for below example:
* max_rank: setting method = 'max' the records that have the same values are ranked using the highest rank (e.g.: since cat and dog are both in the 2nd and 3rd position, rank 3 is assigned.)
* NA_bottom: choosing na_option = 'bottom', if there are records with NaN values they are placed at the bottom of the ranking.
* pct_rank: when setting pct = True, the ranking is expressed as percentile rank.

## Ranking on partition (Similar to rank() by Parition in SQL)

````python
rank_df = pd.DataFrame(
    {
        "group": ["a", "a", "a", "a", "a", "b", "b", "b", "b", "b"],
        "value": [2, 4, 2, 3, 5, 1, 2, 4, 1, 5],
    }
)
````

````python
for method in ['average', 'min', 'max', 'dense', 'first']:
    rank_df[f'{method}_rank'] = rank_df.groupby('group')['value'].rank(method)

rank_df
````

* **groupby()** in Pandas: The groupby() function creates "virtual groups" by splitting the DataFrame into smaller groups based on the values in the 'group' column. It doesn't immediately perform any computation or aggregation. Instead, it creates a "grouped" DataFrame object that can be operated on.

* **Ranking within each group**: When you call the .rank() function on ['value'], it's applied within each of these groups. So instead of aggregating (collapsing) the values, pandas applies the ranking logic separately to each group.

* **No Aggregation Function**: Aggregation functions (like sum(), mean(), min(), etc.) summarize the data into a single value per group. But here, ranking doesn't summarizeit generates a ranking of values within each group. So, after groupby(), the size of the data remains the same because each value is assigned a rank within its respective group.

# Rolling Function


**`monthly_revenue['rolling_avg'] = monthly_revenue['Monthly Revenue'].rolling(window=3, min_periods=1).mean()`**

* This line creates a rolling window object with a window size of 3 over the DataFrame df1.

>**`.rolling(3):`** This specifies that the rolling window will consider 3 consecutive rows of data to calculate statistics (such as the mean).

>**`min_periods=1:`** This ensures that even if fewer than 3 data points are available (e.g., at the start of the dataset), the rolling calculation will still be applied with the available data. For example, if only 1 or 2 values are available, it will compute the mean of those.
Example:

* For a dataset like this:

| year_month | Monthly revenue | rolling_avg  |
|------------|--------------|--------------|
| 2020-01    | 26292        | 26292        |
| 2020-02    | 20695        | 23493.5      |
| 2020-03    | 29620        | 25535.667    |
| 2020-04    | 21933        | 24082.667    |



The rolling window would work as follows:

* For 2020-01, only the first value (26292) is available, so it takes that as the mean.
* For 2020-02, two values (26292, 20695) are available, so it takes the mean of these two.
* For 2020-03, three values (26292, 20695, 29620) are available, and it computes the mean of these three.
* For 2020-04, the rolling window shifts forward to take the last 3 values (20695, 29620, 21933).


# Concat

 `pd.concat()`

The `pd.concat()` function is used to concatenate two or more pandas `DataFrame` or `Series` objects along a particular axis (rows or columns).

 **Syntax:**

```python
pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, ...)


**Common Parameters:**

- **`objs`**: A sequence or mapping of `DataFrame` or `Series` to concatenate.
- **`axis`**: {0, 1}, default is 0. Concatenation direction:
  - `0` (default)  concatenate along rows (vertically).
  - `1`  concatenate along columns (horizontally).
- **`join`**: {outer, inner}, default is 'outer'. Whether to do an inner or outer join:
  - **`outer`**  union of indices (default).
  - **`inner`**  intersection of indices.
- **`ignore_index`**: If True, the resulting axis will be labeled starting from 0. The original index values are ignored.
- **`keys`**: If passed, it creates a hierarchical index using the passed keys.


* **CONCATING** along rows

````python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenating along rows
result = pd.concat([df1, df2], axis=0, ignore_index=False)
print("original indexes are kept \n", result)


result = pd.concat([df1, df2], axis=0, ignore_index=True)
print("indexes are reset \n", result)
````

* **CONCATING** along cols

````python
# Concatenating along cols
result = pd.concat([df1, df2], axis=1)
print(result)
````

### Category Operations

The **`.cat`** accessor is used for working with categorical data types. It provides a way to interact with the categorical values and categories associated with a column.


```python

#Access Categories
df['category_column'].cat.categories

#Rename Categories
category_map = {'old_category': 'new_category'}
df['category_column'] = df['category_column'].cat.rename_categories(category_map)

#Add Categories (It will not add any data in df, but will update new_category with zero records)

# Remove Unused Categories
df['category_column'] = df['category_column'].cat.remove_unused_categories()
  """
    This command removes categories that are defined but are not actually used in the data.

    After creating categorical columns and performing some data operations, it's possible to end up with categories that aren't represented in the actual data. These "unused" categories take up space in the category definition but don't exist in the column itself.
    
    In this case, since ['Very High', 'Very Low'] was added as a category but was never used in the actual data, it gets removed.

    Before: Index(['Low', 'Medium', 'Very High','Very Low'], dtype='object')

    After: Index(['Low', 'Medium'], dtype='object')


  """



# Reorder Categories
df['category_column'] = df['category_column'].cat.reorder_categories(['Low', 'Medium', 'High'], ordered=True)
  '''
    This command reorders the existing categories in a specific sequence and sets the ordered flag to True, meaning that the order is now meaningful for comparisons.

    Categorical columns in Pandas can either be unordered or ordered. In an unordered category, the order of the categories doesn't matter (like 'red', 'green', 'blue'). In an ordered category, there's an inherent order (like 'Low', 'Medium', 'High').

    The .cat.reorder_categories() method changes the order of the categories. When ordered=True is set, Pandas understands that there's a meaningful order to the categories, which allows comparisons between them (e.g., Low < Medium < High)

  '''

```
